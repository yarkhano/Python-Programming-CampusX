## Tuples ##

**Tuples are like the list with only these changes that tuples has small brackets while list has squared and tuples are immutable and lists are mutable**

```python
#creating tuples
t1 = (1,2,3)
```

```python
#a single element in a tuple is not a tuple add a comma after it

t2 = (2)
type(t2)
```

```python
t2 = (2,)
type(t2)
```

**Tuple is not editable**

**Tuples are speedy than and**
**Tuples occupy less space than lists due to immutability**

```python
#tuple unpacking
a,b,c = (1,2,3)

print(b)
```

```python
#variable wapping
a = 1
b = 2
a,b = b,a
print(a,b)
```

```python
#need only first two values

a,b,*others = (1,2,3,4)
print(a,b)
```

```python
#frozensets are immutable sets
fs = frozenset([1,2,3])
fs
```

## Dictionaries ##

**can not add mutable data types inside the dict.key must always be unique.No indexing.Mutable.No duplicate keys**

```python
#Accessing dicts
d1 = {"name":"yar","age":23}
#cant be access using indexing

#two ways to access
d1["name"]
d1.get("name")
```

```python
#adding new pairs
d1["weight"] = 60
d1
```

```python
for i in d1:
    print(i,d1[i])
```
