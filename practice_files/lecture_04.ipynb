list is heterogenous(can store any data type),while array is homogenous.
lists has slow execution and consume more memmory.

Lists is also called refrential array

list is homogenous because it stores pointers or addresses not exact element

Lists are mutable,while trings are not

```
a = [1,2,3]
```

## Creating List ##

```
#3D List
threed = [
    [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ],
    [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ],
]
```

## Accessing List ##

```
print(threed[0][1][1])
```

```
#reverse list
print(threed[::-1])
```

## Adding To List ##

```
l = [1,2,3]
```

```
#append add always single item
l.append(4)
print(l)
```

```
l.append([4,5,6])
print(l)
```

```
#extend add multiple items
s = [1,2,3]
s.extend([7,8,9])
print(s)
```

```
s.extend("Islamabad")
print(s)
```

```
#insertion use for desired position
i = [1,2,3,4]

i.insert(1,5)
print(i)
```

## Deleting ##

```
#delete
dlist = [1,2,3,4,5]
del dlist[1]
print(dlist)
```

```
dlist = [1,2,3,4,5]
del dlist
```

```
#remove use to delete a specific element
rlist = [1,2,3,4,5]
rlist.remove(5)
print(rlist)
```

```
#clear delete the list
clist = [1,2,3,4,5]
clist.clear()
print(clist)
```

```
#pop remove last element if anyone is not mentioned
plist = [1,2,3,4,5]
plist.pop()
print(plist)
```

```
plist = [1,2,3,4,5]
plist.pop(0)
print(plist)
```
