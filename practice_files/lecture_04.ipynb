## Creating List ##

```
#3D List
threed = [
    [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ],
    [
        [1,2,3],
        [4,5,6],
        [7,8,9]
    ],
]
```

## Accessing List ##

```
print(threed[0][1][1])
```

## Adding To List ##

```
l = [1,2,3]
```

```
#append add always single item
l.append(4)
print(l)
```

```
l.append([4,5,6])
print(l)
```

```
#extend add multiple items
s = [1,2,3]
s.extend([7,8,9])
print(s)
```

```
s.extend("Islamabad")
print(s)
```

```
#insertion use for desired position
i = [1,2,3,4]

i.insert(1,5)
print(i)
```

## Deleting ##

```
#delete
dlist = [1,2,3,4,5]
del dlist[1]
print(dlist)
```

```
dlist = [1,2,3,4,5]
del dlist
```

```
#remove use to delete a specific element
rlist = [1,2,3,4,5]
rlist.remove(5)
print(rlist)
```

```
#clear delete the list
clist = [1,2,3,4,5]
clist.clear()
print(clist)
```

```
#pop remove last element if anyone is not mentioned
plist = [1,2,3,4,5]
plist.pop()
print(plist)
```

```
plist = [1,2,3,4,5]
plist.pop(0)
print(plist)
```

## List Comprehension ##

**comprehension is a shortcu way writing multiple items to a list**

```
[i for i in range(1,21)]
```

```
al = [1,2,3,4,5]
[i**2 for i in al]
```

```
basket = ["apple","banana","guava"]
my_fruits = ["orange","apple"]

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```
[[i*j for i in range(1,4) for j in range(1,4)]]
```

```
l1 = [1,2,3,4]
l2 = [5,6,7,8]

[i*j for i in l1 for j in l2]
```

## ZIP() ##

**Zip function create pairs and we can also make a third list using addition of first and second list**

```
l1 = [1,2,3,4]
l2 = [1,2,3,4]

[i+j for i,j in zip(l1,l2)]
```
