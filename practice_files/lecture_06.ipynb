```python
def power(a=2,b=2):
    """This function show default arguments"""
    return a**b
```

**positional arguments -> it is normal or usual order or arguments passes to the function**

**keyword arguments -> where you add arguments by name and there is no positional order and function is executed as inteneded/normal**

**this(starics args)-> use when you want to pass multiple arguments to function**

```python
def multiply(*args):
    product = 1
    for num in args:
     product *=num
    return product
```

```python
multiply(1,2,3,9)
```

**(staric staric kwargs->use for key pair values**

```python
def countries(**kwargs):
    print()
    print("counties and there capitals")
    print()
    for (key,value) in kwargs.items():
        print(key,'->',value)
```

```python
countries(pakistan="islamabad",india="delhi")
```

**Higher order functions-> Function which return another functon or receive anothere as an input**

**Lambda is used with higher order function**

```python
def trans(f,L):
    output = []
    for num in L:
        output.append(f(num))
    return output

trans(lambda x:x**2,[1,2,3])
```

## MAP(higher order function) ##

```python
#map expect a lambda function and an iterateble
l = [1,2,3,4,5,6]
list(map(lambda x: 'even ' if x % 2 == 0 else 'odd', l ))
```

```python
#filter is aslo hof and expect a lambda and and list
l = [1,2,3,4,5,6]
list(filter(lambda num:num>3,l))
```

```python
#reduce -> minimize or add
import functools
functools.reduce(lambda x,y:x+y,l)
```
