**Default arguement**

```python
def power(a=2,b=2):
    """This function show default arguments"""
    return a**b
```

```python
power(3) #second value is not provided
```

**positional arguments -> it is normal or usual order or arguments passes to the function**

**keyword arguments -> where you add arguments by name and there is no positional order and function is executed as inteneded/normal**

**this(starics args)-> use when you want to pass multiple arguments to function**

```python
def multiply(*args):
    product = 1
    for num in args:
     product *=num
    return product
```

```python
multiply(1,2,3,9)
```

**(staric staric kwargs->use for key pair values**

```python
def countries(**kwargs):
    print()
    print("counties and there capitals")
    print()
    for (key,value) in kwargs.items():
        print(key,'->',value)
```

```python
countries(pakistan="islamabad",india="delhi")
```
