# Nested Loops

```python
rows = 5

for i in range(1,rows+1):
    for j in range(1,i+1):
        print("*",end=" ")
    print()
```

```python
rows = 4

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end=" ")
    for k in range(i-1,0,-1):
        print(k,end=" ")
    print()
```

```python
lower = 10
upper = 100

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
     print(i)
```

## Accessing Strings

**positive indexing**

```python
name = 'yar khan'
print(name[0])
```

**Negative Indexing**

```python
name = 'yar khan'
print(name[-1])
```

**Slicing**

```python
name = 'yar khan'
print(name[0:8])
print(name[:8])  # start to 8
print(name[0:])  # start to end
print(name[2:])  # 2 to end
```

```python
print(name[0:6:2])
```

```python
name = 'yar khan'

print(name[-4:-7])
```

```python
#to reverse string
print(name[::-1])
```
