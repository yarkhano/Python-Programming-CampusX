# Nested Loops

```python
rows = 5

for i in range(1,rows+1):
    for j in range(1,i+1):
        print("*",end=" ")
    print()
```

```python
rows = 4

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end=" ")
    for k in range(i-1,0,-1):
        print(k,end=" ")
    print()
```

```python
lower = 10
upper = 100

for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
     print(i)
```

```python
name = "سلام"
print(ord(name[0]))
```
