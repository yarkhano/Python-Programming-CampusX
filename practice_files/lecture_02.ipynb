# Operators

```python
print(5//2)  #integer division
```

```python
#logical operators
print(1 and 1)
```

```python
#bitwise operators
print(2&3)
print(2|3)
print(2^3)

#left shift
print(5>>2)

#right shift
print(5<<2)
```

```python
#Memebership Operators
#in/not in

print('y' in 'yar')
print(1 in [2,3,4,5])

print(2 not in [3,4,5])
```
